```python
# feature selection for logistic regression on the whole dataset

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression



# Train a logistic regression model with feature selection
selector = RFE(LogisticRegression(), n_features_to_select= 23)
selector = selector.fit(X, y)


# Print the selected feature names and whether they are supported
print(X.columns[selector.support_])
print(selector.support_)

```

```python
Features_selected = X.columns[selector.support_]
```

```python
# data imbalancing - SMOTE on train set


from imblearn.over_sampling import SMOTE
from collections import Counter

counter = Counter(y_train)
print('Before', counter)

smt = SMOTE()
x_train_smt, y_train_smt = smt.fit_resample(X_train, y_train)
counter = Counter(y_train_smt)
print('After', counter)

```

```python
print(x_train_smt)
```

```python
# logistic regression

from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(x_train_smt[Features_selected], y_train_smt)
```

```python
# Predict labels for instances in testing set
y_pred = clf.predict(X_test[Features_selected])
```

```python
# Compute accuracy, precision, recall, and f1 score using skleanr.metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1:', f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf, X_test[Features_selected], y_test)  
plt.show()  
```

```python
# since we do not know the ideal numbers of features for our model we run a for loop from a range of numbers to find out
def metrics_features(num_features):
    selector = RFE(LogisticRegression(), n_features_to_select = num_features)
    
    
    selector = selector.fit(X_train, y_train)
    
    Features_selected = X_train.columns[selector.support_]
    
    clf = LogisticRegression(random_state = 1).fit(x_train_smt[Features_selected], y_train_smt)
    
    y_pred = clf.predict(X_test[Features_selected])
    
    Accuracy = accuracy_score(y_test, y_pred)
    
    Precision = precision_score(y_test, y_pred)
    
    Recall = recall_score(y_test, y_pred)
    
    F1 = f1_score(y_test, y_pred)
    
    return Accuracy, Precision, Recall, F1
features_num = []    
accuracy = []
precision = []
recall = []
f1 = []
for i in range(3,29):
    metrics = metrics_features(num_features = i)
    accuracy.append(metrics[0])
    precision.append(metrics[1])
    recall.append(metrics[2])
    f1.append(metrics[3])
    features_num.append(i)
    
    
```

```python
print(accuracy, precision, recall, f1 )
```

```python
# Visualizing the results according to the number of selected features
import matplotlib.pyplot as pyplot

pyplot.rcParams['figure.figsize'] = (16, 8)
pyplot.rcParams['font.size'] = 15
pyplot.plot(features_num, accuracy, label = 'accuracy')
pyplot.plot(features_num, precision, label = 'precision')
pyplot.plot(features_num, recall, label = 'recall')
pyplot.plot(features_num, f1, label = 'f1')
pyplot.legend()
pyplot.xlabel('number of feature selected')
pyplot.ylabel('values')

```

```python
#performing statistical test to check the difference in the data
import pandas as pd
import scipy.stats as stats
```

```python
from scipy.stats import ttest_ind, ranksums

churn_tenure = df[df['Churn']==1]['Tenure']
not_churn_tenure = df[df['Churn']==0]['Tenure']

# Wilcoxon rank-sum test
wilcoxon_statistic, p_value = ranksums(churn_tenure, not_churn_tenure)
print("Wilcoxon p-value for tenure: ", p_value)
```

```python
from scipy.stats import chi2_contingency

citytier_churned = pd.crosstab(df['CityTier'], df['Churn'], margins=True)
chi2, p_value, dof, expected = chi2_contingency(citytier_churned)
print("Chi-square p-value for CityTier: ", p_value)
```
